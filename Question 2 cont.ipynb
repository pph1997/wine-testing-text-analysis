```python
import gradio as gr
import pandas as pd
import plotly.express as px

# Load your cleaned data (assuming the Excel file is in the same directory)
df = pd.read_excel("Word list cleaned.xlsx")

# Preprocess: group top keywords by country
keywords_per_country = (
    df.groupby("country")["Revamp words"]
    .apply(lambda x: ", ".join(x.head(5)))
    .reset_index(name="Top Keywords")
)

# Define function to return interactive plotly figure
def plot_world_keywords():
    fig = px.choropleth(
        keywords_per_country,
        locations="country",
        locationmode="country names",
        color_discrete_sequence=["#636EFA"],
        hover_name="country",
        hover_data={"Top Keywords": True, "country": False},
        title="Top Keywords by Country"
    )
    fig.update_geos(showcountries=True, projection_type="natural earth")
    fig.update_layout(margin={"r":0,"t":50,"l":0,"b":0})
    return fig

# Gradio Interface
demo = gr.Interface(fn=plot_world_keywords, inputs=[], outputs=gr.Plot(), title="Interactive World Keyword Map")
demo.launch()
```
